### Q: "차 무게가 무거울수록 연비가 낮을까?"
- 이런 가설을 세우고 증명하려면 어떻게 해야해?
- 일단 h0 = 차가 무거울수록 연비가 낮다 , h1= 무게와 관련이 없다
- 라고 잡고 weigth랑 horsepower랑 비교하면 되려나?

🚗 질문:
"차 무게가 무거울수록 연비가 낮을까?"

→ 이건 음의 상관관계가 있는지 보는 거야.
-  1단계: 가설 세우기
    - 귀무가설 H₀: 차 무게와 연비는 관련이 없다. (상관관계 없음)

    - 대립가설 H₁: 차 무게가 무거울수록 연비가 낮다. (음의 상관관계 있음)

```python
###2단계: 데이터 준비
#우리는 mpg와 weight 열을 사용할 거야.
mpg[['mpg', 'weight']].head()
```

```python
### 3단계: 시각화로 먼저 확인
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(data=mpg,x='weight',y='mpg')
plt.title("차 무게 vs 연비")
plt.show()
```

```python
#음의 상관관계를 띄는걸 볼 수 있음 즉, h1 채택 ho 기각
```

```python
###  4단계: 수치로 확인 (상관계수(r))
corr = mpg[['mpg','weight']].corr()
print(corr)
```

```python
#상관계수(r) : 음의 상관관계
```

###상관계수 (Correlation coefficient) 값:

- -1이면 완벽한 음의 상관관계

- 0이면 무관

- +1이면 완벽한 양의 상관관계

→ 만약 -0.8 같은 값이 나오면 무게가 무거울수록 mpg가 낮다는 걸 보여줘.



```python
### 5단계: 통계적 검정 (선택사항)
from scipy.stats import pearsonr

corr_coef,p_value = pearsonr(mpg['weight'],mpg['mpg'])
print("상관계수:",corr_coef)
print("p값",p_value)
```

## 분석결과
- 상관게수(r): 음의상관관계
- p값 : 유의수준 0.05> 2.9727995640497043e-103 p값
-> 유의수준이 더 크므로 h0 기각
-> 즉, 차의 무게가 증가하면 연비가 낮아진다는 가설은 통계적으로 유의미하다고함



- p < 0.05면 귀무가설 기각 → 상관관계 있다!

- corr_coef < 0 → 음의 상관관계!

### ✅ 3단계: 통계적 차이 검정 (ANOVA 또는 Kruskal)
- 만약 단순 평균 차이뿐만 아니라 통계적으로 유의미한지 보고 싶다면:
- 📌 방법 1: 정규성 가정 → ANOVA

```python
### 📌 방법 1: 정규성 가정 → ANOVA
from scipy.stats import f_oneway

usa = mpg[mpg['origin'] == 'usa']['mpg']
japan = mpg[mpg['origin'] == 'japan']['mpg']
europe= mpg[mpg['origin'] == 'europe']['mpg']

f_stat,p_val = f_oneway(usa,japan,europe)
print("F값:",f_stat)
print("p값:", p_val)
### 유의수준 0.05 > p_value여서 정규성 불충족 -> Kruskal-Wallis test시행
```

💡 하나씩 뜯어서 설명

① from scipy.stats import f_oneway
scipy.stats 모듈에서 f_oneway 라는 함수를 가져옴.

이 함수는 ANOVA(일원분산분석) 를 해줌.

목적: 세 그룹 이상 간의 평균 차이가 통계적으로 유의미한가 확인.

② usa = mpg[mpg['origin'] == 'usa']['mpg']
mpg라는 DataFrame에서 origin이 'usa'인 행만 추출 → 그 중에서도 'mpg'값만 가져옴.

즉, 미국산 차의 연비 데이터만 추출한 시리즈가 됨.

같은 방식으로 japan, europe도 해당 국가의 연비(mpg) 데이터만 뽑은 거야.

③ f_stat, p_val = f_oneway(usa, japan, europe)
f_oneway() 함수에 3개의 연비 데이터를 넣음.

결과로:

f_stat: F-통계량 (각 그룹 간 평균 차이를 나타냄)

p_val: p-값 (차이가 통계적으로 유의미한지 판단하는 값)

④ print("F값:", f_stat)
f 통계량 출력
→ 값이 클수록 그룹 간 평균 차이가 큼.

⑤ print("p값:", p_val)
이게 핵심!
→ p값 < 0.05면 H₀ 기각 → origin에 따라 평균 연비가 다르다고 결론 낼 수 있음.



🔍 정규성 검정 vs 아노바 검정은 완전히 다른 것이에요.
검정 종류	목적
✅ 정규성 검정	데이터가 정규분포를 따르는지 확인 (예: Shapiro-Wilk test, Kolmogorov-Smirnov test 등)
✅ 아노바(ANOVA) 검정	집단 간 평균 차이가 있는지 확인
-----------------------------------------------------------------------
✅ 그래서 정규성 검정을 하기 위해 ANOVA를 사용한 건 아니고요,
정규성 검정을 먼저 한 다음,
정규성을 만족하면 → ANOVA 사용
정규성을 만족하지 않으면 → Kruskal-Wallis test 같은 비모수 검정을 사용합니다.

예: shapiro() 함수를 써서 각 그룹(mpg[origin==USA], mpg[origin==Japan], ...)에 대해 확인

- 정규성 충족 → ANOVA
- 정규성 불충족 → Kruskal-Wallis test

- 검정 결과에 따른 p-value 해석 → H0 기각 여부 판단

```python
###📌 방법 2: 정규성 없는 경우 → Kruskal-Wallis
from scipy.stats import kruskal

h_stat, p_val = kruskal(usa,japan,europe)
print("H값:",h_stat)
print("p값:", p_val)
```

✅ 해석 방법


- p값 < 0.05: 지역(origin)에 따라 연비 평균 차이가 통계적으로 유의미함

- p값 ≥ 0.05: 유의미한 차이가 없음

- 정규성이 있는 경우 없는경우 모두
- 유의수준 0.05 > p-value 니까 H0기각 h1 채택
-> 즉, origin에 따라 평균 연비 차이가 있다는거임
