### 'IterativeImputer'의 주요 하이퍼파라미터(hyperparameter)

1. estimator : 누락된 값을 추정하기 위해 사용할 추정기. 기본값은 'BayesianRidge()'이며, 'LInearRegression','DecisionTreeRegressor','KNeighborsRegressor' 등 다른 회귀 추정기로 변경할 수 있음 

2. max_iter : 최대 반복 횟수로, 누락된 값을 추정하는 최대 사이클 수. 기본값은 10임 

3. random_state: 난수 생성기의 시드 값을 설정함. 동일한 결과를 재현하고 싶을 때 유용함 

4. initial_stratety : 누락된 값을 초기에 대치할 전략으로, 'mean','median','most_frequent',또는 'constant' 중 하나를 선택할 수 있음. 기본값은 'mean'dla 

5. imputation_order : 특성을 대치하는 순서를 결정함. 'ascending','descending','roman','arabic',또는 'random' 중에서 선택할 수 있음. 기본값은 'ascending'임 

6. skip_complete : True로 설정하면, 완전한 케이스(누락된 값이 없는 경우)를 계산에서 제외함. 기본값은 False임.

7. add_indicator: 누락된 특성에 대한 추가적인 지표(누락된 값이 있음을 나타내는 이진 변수)를 추가할 지 여부를 결정함. 기본값은 False임 

8. tol : 반복을 멈출 임계값으로, 이전과 현재 반복 사이의 평균 제곱 차이가 이전값보다 작을 경우 반복이 중지됨. 기본값은 '1e-3'(	1 × 10⁻³,0.001) 임 

9. n_nearest_features : 각 특성을 대치할 때 고려할 다른 특성의 수임. None으로 설정하면 모든 특성이 사용됨. 기본값은 None임 

10. verbose : 실행 중 출력 메시지의 양을 제어함. 값이 클수록 더 많은 메시지가 출력됨. 기본값은 '0'임 

11. sample_posterior : True롤 설정하면, BayesianRidge 추정기를 사용할 떄 샘플링을 통해 후방 분포에서 값을 추정함. 기본값은 False임 

### SimpleImputer 하이퍼파라미터 : 

1. missing_values : 누락된 값을 나타내는 placeholder임. 기본값은 'np.nan'임. 숫자나 NaN 외에도 다른 표현을 사용할 수 있음

2. strategy : 누락된 값을 대체할 방법을 결정함. 가능한 값은 "mean","median", "most_frequent","constant"임. 기본값은 "mean임 

3. fill_value : 'strategy="constant"'를 사용할 때 누락된 값을 대체하기 위한 상수 값을 설정함. 기본값은 'None'임. 이 값을 설정하면 모든 누락된 값을 해당 값으로 대체함 

4. copy : 'True'인 경우, 원본 데이터는 변경되지 않고 새 데이터 배열이 생성됨. 'False'일 경우, 원본 데이터가 대체됨. 기본값은 'True'임 

5. add_indicator : 'True'인 경우, 누락된 값이 있는 특성에 대한 indicator(지시자) 변수가 결과에 추가됨. 기본값은 'False임 



### 'KNNImputer' 하이퍼파라미터 :

```python
## 모델에 대한 성능을 담을 평가 데이터프레임

results = pd.DataFrame(columns = ['Method','Dataset','Accuracy','Precision','Recall','F1'])
    # 코드는 results라는 빈 데이터프레임(DataFrame)을 만드는 코드인데,
    # 앞으로 머신러닝 모델의 성능 평가 결과를 정리해서 담을 때 자주 쓰임
    #pd.DataFrame()	,새로운 데이터프레임을 만든다
    #columns=[...],	이 데이터프레임의 **열 이름(컬럼)**을 지정
    #results	만들어진 데이터프레임을 저장할 변수 이름


#결측치 모두 제거 
def drop_na(df):
    return df.dropna(subset=['age']) #컬럼을 지정할 수 있음
    # age 컬럼에 결측치(NaN)가 있는 행만 제거하겠다는 뜻

#결측치를 평균으로 대치 
def mean_imputation(df):
    df_filled = df.copy()
    df_filled['age'].fillna(df_filled['age'].mean(), inplace=True)
    return df_filled
    #df.copy()	, 원본 데이터프레임을 변경하지 않기 위해 복사본을 생성해요.
    # df_filled['age'].mean(), 	age 컬럼의 평균값을 계산해요. (결측치는 자동 제외됨)
    # .fillna(..., inplace=True), 	결측치를 지정한 값으로 바로 채움. 여기선 age에 있는 NaN을 평균으로 바꿈
    # return df_filled , 	결측치를 평균으로 채운 새로운 데이터프레임을 반환

#선형보간으로 
def linear_interpolation(df):
    df_filled = df.copy()
    df_filled['age'].interpolate(method='linear',inplace=True)
    return df_filled

# 코드	설명
# df.copy()	원본을 손상시키지 않기 위해 복사본 생성
# df_filled['age'].interpolate(...)	결측치를 보간해서 채움. 기본은 선형보간(linear)
# method='linear'	NaN 앞뒤의 숫자들을 기준으로 일정한 간격으로 값 채우기
# inplace=True	바로 해당 열을 바꿔서 저장 (새로운 객체 안 만들고 바로 덮어쓰기)
# return df_filled	보간된 결과를 반환
## 선형보간은, 결측치 앞뒤의 값을 직선으로 이어서,
## 그 사이에 있는 NaN을 그 선 위의 값으로 예측해서 채우는 방식이야.



#KNN으로 진행
def knn_imputation(df,n_neighbors =5):
    df_filled = df.copy()
    imputer=KNNImputer(n_neighbors = n_neighbors)
    df_filled[['age']]=imputer.fit_transform(df_filled[['age']])
    return df_filled

#코드	설명
# df.copy()	, 원본 데이터를 보호하기 위해 복사
# KNNImputer(n_neighbors=n_neighbors)	,KNN 기반 결측치 대체 도구 생성 (sklearn에서 제공)
# imputer.fit_transform(df_filled[['age']]),	'age' 컬럼의 NaN 값을 K개의 이웃 데이터를 기반으로 예측해서 채움
# df_filled[['age']] = ...	,채워진 값을 다시 'age' 컬럼에 대입
# return df_filled	, 완성된 데이터프레임 반환

#다양한 방법으로 결측치를 보간하고 RF 모델에 학습시켜 train,test의 생존율 예측 결과를 확인하자!

# 각 방법을 적용하여 성능 테스트
methods = {
    'Drop NA': drop_na,
    'Mean Imputation':mean_imputation,
    'Linear Interpolation':linear_interpolation,
    'Knn Imputation' : knn_imputation
}

## 타이타닉 데이터로 반복을 돌려서 방법들을 다 하나씩 돌려서 성능 테스트를 넣어야 한다.
## 데이터를 학습시키기 위해서 train, test 로 나눠야 한다.
## 인코딩하는 get_dummies 인코딩 작업

for method_name, impute_func in methods.items():
    titanic_filled = impute_func(titanic)

# 코드	의미
# methods	dict 형태로 만들어진, 여러 결측치 처리 방법 모음 (예: 평균 대체, 선형 보간, KNN 등)
# .items()	딕셔너리에서 (key, value) 쌍을 하나씩 꺼내주는 함수
# for method_name, impute_func in ...	반복문: key는 함수 이름 ('mean', 'linear' 등), value는 해당 함수 (mean_imputation, linear_interpolation 등)
# impute_func(titanic)	titanic 데이터를 함수에 넣어서 결측치를 채운 새로운 데이터프레임 생성
# titanic_filled	결측치를 처리한 결과가 저장되는 변수 (한 번에 하나씩)
    
    #학습할 데이터셋으로 나누기
    X = titanic_filled.drop(['survived','deck','alone','who','adult_male','alive','embark_town'],axis = 1)
    y = titanic_filled['survived']
    
    #원핫 인코딩으로 범주형 변수를 수치로 바꾸는 것
    X = pd.get_dummies(X, drop_first=True)    
    
    #데이터셋 분할
    X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.2, random_state=111)
    
    model = RandomForestClassifier(random_state=111)
    model.fit(X_train, y_train)
    
    # 성능에 대해서 평가한 것들을 데이터프레임 저장해야 한다.
    
    for dataset, X_data, y_data in (['Train',X_train, y_train],('Test',X_test, y_test)):
        '''
        for dataset, X_data, y_data in ... 부분은 훈련 데이터(X_train, y_train)와 
        테스트 데이터(X_test, y_test)에 대해 반복문을 실행하는 코드입니다.

        ['Train', X_train, y_train]와 ['Test', X_test, y_test]는 각각 훈련 데이터와 
        테스트 데이터를 반복하면서 처리하도록 되어 있습니다.

        dataset은 'Train' 또는 'Test'가 되며, X_data와 y_data는 각각 
        특성 데이터(X_train, X_test)와 목표 변수(y_train, y_test)를 의미합니다.
        '''
        y_pred = model.predict(X_data)
        '''
        model.predict(X_data)는 훈련된 모델을 사용하여 
        **특성 데이터(X_data)**에 대해 예측을 수행합니다.

        y_pred는 예측된 값으로, 실제 값 y_data와 비교하여 모델의 성능을 
        평가하게 됩니다.
        '''
        # 성능 지표 계산
        #'Accuracy','Precision','Recall','F1'
        accuracy = accuracy_score(y_data, y_pred)
        precision = precision_score(y_data, y_pred)
        recall = recall_score(y_data, y_pred)
        f1 = f1_score(y_data, y_pred)   
        '''
        accuracy_score(y_data, y_pred): 정확도를 계산. 
        예측값과 실제값이 얼마나 일치하는지를 나타냄

        precision_score(y_data, y_pred): 정밀도를 계산합니다. 
        모델이 양성 클래스라고 예측한 것들 중 실제로 양성인 것의 비율

        recall_score(y_data, y_pred): 재현율을 계산합니다. 
        실제 양성 중에서 모델이 양성이라고 정확하게 예측한 것의 비율

        f1_score(y_data, y_pred): F1-score를 계산합니다. 
        정밀도와 재현율의 조화 평균으로, 두 지표의 균형을 고려한 성능 평가 지표
        '''      
        
        #결과 저장
        result_now=pd.DataFrame({
            'Method':[method_name],
            'Dataset':[dataset],
            'Accuracy':[accuracy],
            'Precision':[precision],
            'Recall':[recall],
            'F1':[f1]
        })
        results = pd.concat([results, result_now], ignore_index=True)
        
print(results)
```

## 모델 성능 평가 결과 해석

이 결과는 여러 결측치 처리 방법에 대해 모델의 성능을 평가한 것입니다. 평가된 성능 지표는 **정확도(Accuracy)**, **정밀도(Precision)**, **재현율(Recall)**, **F1-score**입니다. 각 성능 지표는 모델이 얼마나 잘 동작하는지를 측정하는 데 중요한 역할을 합니다.

---

### 1. **데이터셋에 따른 성능 차이**
- **훈련 데이터(Train)**와 **테스트 데이터(Test)**에 대한 성능을 비교할 수 있습니다.
  - 훈련 데이터에서는 모델이 **상대적으로 높은 성능**을 보이는 경우가 많습니다. 이는 모델이 훈련 데이터에 대한 패턴을 잘 학습했기 때문입니다.
  - 테스트 데이터에서는 **성능이 더 낮을 수 있습니다**. 이는 모델이 훈련 데이터에 과도하게 적합(overfitting)되어 일반화 능력이 떨어질 수 있음을 시사합니다.

---

### 2. **결측치 처리 방법에 따른 성능 차이**
- **Drop NA** (결측치 삭제):
  - 훈련 데이터에서 성능이 **매우 높지만**, 테스트 데이터에서는 **상대적으로 낮은 성능**을 보였습니다. 결측치를 삭제하면 데이터 손실이 발생해, 테스트 데이터에 대해 성능이 떨어질 수 있습니다.
  
- **Mean Imputation** (평균 대체):
  - 훈련 데이터에서 **높은 성능**을 보였지만, 테스트 데이터에서 **조금 낮은 성능**을 보였습니다. 평균값 대체는 간단하지만, 데이터의 분포를 왜곡할 수 있습니다.
  
- **Linear Interpolation** (선형 보간법):
  - 훈련 데이터에서 **상당히 높은 성능**을 보였고, 테스트 데이터에서도 **비교적 높은 성능**을 유지했습니다. 선형 보간법은 데이터 간의 패턴을 잘 반영할 수 있어 성능을 개선할 수 있습니다.
  
- **KNN Imputation** (K-최근접 이웃):
  - 훈련 데이터에서는 **평균 대체 방법**과 비슷한 성능을 보였고, 테스트 데이터에서는 **조금 낮은 성능**을 보였습니다. KNN은 주변 데이터의 패턴을 반영하지만, 데이터의 분포에 따라 성능이 달라질 수 있습니다.

---

### 3. **모델 성능 지표 해석**
- **Accuracy (정확도)**: 
  - 훈련 데이터에서 모델의 정확도는 **높게 나타났으며**, 테스트 데이터에서는 **Drop NA**와 **KNN Imputation**이 상대적으로 낮은 정확도를 보였습니다.

- **Precision (정밀도)**:
  - **테스트 데이터에서 정밀도가 낮은 경향**을 보였습니다. 이는 모델이 양성이라고 예측한 것 중 실제 양성인 것의 비율이 떨어질 수 있음을 나타냅니다.

- **Recall (재현율)**:
  - **재현율 또한 테스트 데이터에서 낮은 값**이 나타났습니다. 이는 모델이 양성 클래스의 데이터 중 실제 양성을 잘 잡아내지 못하는 경우가 있다는 것을 의미합니다.

- **F1-score**:
  - **정밀도와 재현율**의 조화 평균인 F1-score도 훈련 데이터에서는 **높은 값**을 보였으나, **테스트 데이터에서 다소 떨어진 경향**이 있습니다. 이는 모델이 양성과 음성을 균형 있게 예측하는 데 어려움을 겪을 수 있음을 나타냅니다.

---

### 4. **결론 및 해석**
- **Linear Interpolation**은 훈련 데이터와 테스트 데이터에서 모두 **상위 성능**을 보였고, 결측치 처리에서 **좋은 결과**를 나타냈습니다. 선형 보간법은 데이터 간의 관계를 잘 반영할 수 있어, 성능 향상에 기여할 수 있습니다.
  
- **Drop NA** 방법은 **테스트 성능이 상대적으로 낮은 결과**를 보였으며, 결측치가 많을 경우 데이터 손실이 성능 저하를 일으킬 수 있습니다.
  
- **KNN Imputation**은 **훈련 성능**은 좋은 편이지만, **테스트 성능에서 떨어진 경향**을 보였습니다. KNN은 주변 데이터와의 유사성을 기반으로 결측치를 채우기 때문에, 데이터의 특성에 따라 성능 차이가 발생할 수 있습니다.

---

### 5. **추천**
- **Linear Interpolation** 방식이 훈련과 테스트 성능에서 **균형 잡힌 성능**을 보였기 때문에, 결측치 처리 방법으로 **가장 적합**할 수 있습니다.
- **Drop NA** 방법은 결측치가 많을 경우 **피하는 것이 좋습니다**. 데이터 손실이 성능에 큰 영향을 미칠 수 있습니다.

```python
##  단순하게 결측치를 제거하는 것 보다는 어떤 도메인 로직을 통해서 
## 결측치를 보간하는 방법도 있지 않을까? 

titanic[titanic['age'].isna()]

# 요금,class,성별이 중요한 생존에 영향을 주었음

def fill_age_based_on_rules(df):
    #요금을 살펴봐야 하는데 요금을 qxut 4분위로 세분화
    df['fare_bin'] = pd.qcut(df['fare'],q=4,labels=['low','mid-low','mid-high','high'])

    #성별, 등급, 요금의 구간별 평균 나이를 계산 할 수 있음 
    age_means=df.groupby(['sex','pclass','fare_bin'])['age'].mean()

    #결측치를 보간할 때
    def fill_age(row):
        if np.isnan(row['age']):
            try:#우리가 원하는 결측치 보간
                return age_means[row['sex'],row['pclass'],row['fare_bin']]
            except KeyError:
                return df['age'].mean() #해당 그룹이 없으면 전체 평균으로 하고
            
    df['age'] = df.apply(fill_age,axis=1)
    df.drop(columns = ['fare_bin'],inplace=True)

    return df
```
