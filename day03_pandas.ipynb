### Groupby
- 데이터 집계할 떄 이용
- 데이터를 묶고 -> 내가 원하는 통계치를 바라볼 떄 사용
- 남성 4
- 여성 2
- 남성 1
- 여성 7
- 남성 4
- groupby
    - 기준컬럼을 잡고 (성별)
    - 남성 - 합계(9), 평균(3)
    - 여성 - 합계(9), 평균(4.5)
- 데이터프레임.groupby('기준컬럼')
- groupby와 함께 사용하는 agg함수 - 통계치를 빠르게 확인할 때 사용

```python
pc_gr=tt.groupby('pclass')
```

```python
#그룹이 어떤식으로 묶이는지

for i,j in pc_gr:
    print(i,j)

# 그룹별 통계 계산/분석할때, 글룹별로 다른 처리 적용할때,
# 그룹별 시각화,조건별 로직처리할 때 이렇게 사용
```

```python
#groupby를 한 후에 1번으로 값을 넣으면 내가 그룹한 값들의 데이터프레임 반환
pc_gr.get_group(2)
# 이 코드는 tt라는 DataFrame을 pclass로 묶고,
# 그중에서 pclass == 2인 그룹만 뽑아서 반환해.

# 즉, 2등석 승객들만 모은 작은 DataFrame이 나옴
# pclass를 확인해보면 2등급인 승객들만 나온걸 확인할 수 있음
```

```python
#2개 이상도 묶을 수 있음
ps_gr=tt.groupby(['pclass','sex'])
```

```python
for i,j in ps_gr:
    print(i,j)

# 이 코드는 pclass랑 sex 두 개의 열을 기준으로 그룹을 만든 거야.
# 예를 들어서 “1등석 남자”, “1등석 여자”, “2등석 남자”
# … 이런 식으로 **다중 그룹화 (Multi-index GroupBy)**가 되는 거지.

```

```python
ps_gr.get_group((2,'female'))
#pclass가 2등급이고 female인 사람들을 기준으로 정렬
#get_group() vs groupby()
#groupby() → 묶을 준비
#get_group() → 묶인 것 중 하나를 꺼냄
#ps_gr = tt.groupby(['pclass', 'sex'])
    #pclass와 sex 기준으로 데이터를 묶을 준비만 한 상태
#ps_gr.get_group((2, 'female'))
    #pclass가 2등급이고 성별이 female인 data를 꺼집어내서 보는거임
# groupby는 관련해서 그룹=폴더를 만드는 것이고
# get_group은 폴더안에 data를 꺼내보는 것과 같음
```

```python
##groupby를 한 후 특정 통계치를 보고 싶은 경우
# sum, mean, mediam ,...
# agg(새로운 컬럼명=(기준컬럼,통계치))...n증가
tt.groupby('pclass').agg(fare_sum=('fare','sum'))
#pclass로 묶고 등급별 fare를 합계낸거임
```

```python
#na값은 제외하고 계싼됨
tt.groupby('pclass').agg(fare_sum=('fare','sum'),
                         fare_mean=('fare','mean'),
                         age_mean=('age','mean'),
                         age_cnt=('age','count'))
```

```python
#특정 함수를 만들어서 적용 가능함
def min_max(x):
    return x.max()-x.min()
```

###qcut,cut
- qcut:데이터를 분위수(Quantitle) 따라 구간을 나눈다
    - 분위수 기준으로 균등하게 데이터를 나누는 함수
    - pd.qcut(x,q,labels=???, rertbins=??,precision=?,duplicates=??)
    - x는 구간 데이터로 나눌 시리즈 형태 또는 배열
    - q 구간 나눌 분위수 e.g 4면 4분위수
    - labels 각 구간에 붙일 이름
    - retbins 구간의 경계를 함께 반환할지
    - precision 구간의 경계의 소숫점 자릿수
- cut : 고정된 값의 구간으로 나누는 함수, 데이터의 값 범위에 따라 일정한
 크기의 구간을 정의하고, 각 데이터가 해당하는 구간을 속하게 만들면 됨
    - pd.cut(x,bins,right=???,나머지 동일,include_lowest=가장 낮은 구간 포함할지 여부)
    - 위와 동일한 hyperparameter


```python
## qcut 예시 데이터
import numpy as np

data = np.random.randn(10)
print(data)
#np.random.randn(10)
#→ 정규분포(평균=0, 표준편차=1) 를 따르는
#→ 실수형 랜덤값 10개를 생성하는 코드야!
#andn ← 이 함수가 정규분포를 따르는 난수 생성 함수야!
# randn = "random normal" 의 줄임말

# 정규분포란? 평균 0, 표준편차 1인 분포

```

```python
res_qcut=pd.qcut(data, q=4,
                 labels=['1사분위','2사분위','3사분위','4사분위'])
print(res_qcut)

# pd.qcut(data, q=4)
# → data를 4개의 분위(quantile) 로 나눔
# → 즉, 같은 개수로 4등분해서 나눔 (데이터 분포 기준)

# labels=['1사분위','2사분위','3사분위','4사분위']
# → 각 구간에 사람이 보기 편한 이름(라벨) 을 붙임
```

```python
# cut의 예시
data =np.random.randn(10)*100
print(data)
```

```python
res_cut=pd.cut(data,bins=4,labels=['낮음','중간','높음','매우높음'])
print(res_cut)
```

#2. fare 구간을 나눈기
- qcut, cut 나누기

- 구간화를 범주로 만들어서 파생변수를 만들고 확인해보기

- 데이터프레임에 특정 함수를 전체에 적용시킬 때 apply()
- apply(내가 전처리할 함수)

- age에 대한 함수 만들기

```python
def age_cat(age):
    if pd.isnull(age):
        return 'Unknown'
    elif age < 19:
        return 'Child'
    elif 19 <= age <70 :
        return 'Adult'
    else:
        return 'Senior'
```

```python
#apply 함수를 이용해서 쉽게 전처리 할 수 있음
tt['age_group']=tt['age'].apply(age_cat)
```

```python
#fare 요금에 대한 전처리
tt['fare_qcut'] = pd.qcut(tt['fare'],q=4,labels=['1사분위','2사분위','3사분위','4사분위'])
tt['fare_cut'] = pd.cut(tt['fare'],bins=4,labels=['1사분위','2사분위','3사분위','4사분위'])
# quct 에서는 q
# cut에서는 bins
```

### 데이터 병합
- 붙이는 것
    - 왜 붙일까?
    - 기업에서 데이터는 다 DB안에 테이블로 나눠서 보관
    - ML 전처리 데이터 병합
- concat
    - 덩어리와 덩어리를 붙이는 것
    - 데이터프레임 + 데이터프레임을 그대로 붙이는 것
    - df1, df2
    - 붙이는 방법이 2가지
        1. 행 - 가로로 붙이는 방법
        2. 열 - 세로로 붙이는 방법
- merge
    - 공통된 컬럼을 가지고 병합을 하는 것
    - join : 공통 컬럼 기준을 지정한 후 inner, left,right,outer 조인의 연산 방법
    - join , 병합할 때 자세히 설명

```python
test1 = pd.DataFrame({'학회원분반번호':[1,2,3,4,5],
                     '학회원분반':['해바라기반','장미반','튤립반','잡초반','나무반']})
test2 = pd.DataFrame({'학회원분반번호':[2,4,6,8,10],
                      '학회원수':[100,200,300,400,500]})
```

```python
display(test1)
display(test2)
```

```python
pd.merge(test1,test2,how='inner',on='학회원분반번호')

# 🔀 pd.merge(test1, test2, how='inner', on='학회원분반번호')
# 👉 해석:
# on='학회원분반번호'
# → 공통되는 컬럼(조인 기준 컬럼)은 "학회원분반번호"

# how='inner'
# → 공통되는 키 값만 남긴다 (양쪽에 모두 있는 row만)
```

```python
pd.concat([test1,test2],axis=0)

# axis=0은 위아래로(row-wise) 붙이는 거고,

# axis=1은 좌우로(column-wise) 붙이는 거야.

```

```python
pd.concat([test1,test2],axis=1)
```
