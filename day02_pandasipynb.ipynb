- 데이터프레임은 원본데이터를 보존하려는 성질이 있따보니 바로 바뀌지 않음
- 원본 데이터를 바꾸려면 3가지를 생각해야 함
    1. 원본데이터 변수명에 그대로 원본데이터를 넣으면 됨. 변수 사용해서 변수교체
    2. 바로 문법에서 inplace=True 대치를 바로 진행하는 경우
    3. copy() 원본데이터를 카피하여서 실제 원본데이터는 그대로 두고, 내가 분석할 데이터만 변경하여 사용
    
    
    why?
    1. 어느 순간 데이터를 분석하다보면 뭐가 원본이고 무엇을 하고 있는지 가끔 혼선이 올 경우
    2. A로 전처리했는데 B로 전처리해 함. 근데 원본이 없다? 충격과 공포
    3. 이런 여러 예외 케이스를 대응하기 위해 원본데이터는 그대로 두는게 처음부터 좋음

```python
bda_class=pd.DataFrame({
    '이름':['김민지','이조아','김정윤','김찬규','이또아'],
    '상점':[10,20,30,40,50],
    '벌점':[1,2,3,4,5]
    })
```

```python
bda_class
```

```python
# 상점 - 벌점 = 최종상태 컬럼
# 최종상태 변수에 값을 넣으려고 함
```

```python
#변수처럼 사용할 수 있음
bda_class['최종상태'] = bda_class['상점'] - bda_class['벌점']
```

## 데이터 전처리, 추출 등에 필요한 기초 문법

- 성별이 여자인 경우 또는 성별이 남자인 경우

    - query() , 데이터 추출
    - df.loc/iloc[불림참값] , 데이터 추출
    - df[] , 열 기준 추출
    - sort_values(), 정렬

```python
#성별이 남성인 경우
#query (특정 요구조건을 적으면 됨)
## 문자열 데이터는 같은지 여부로 데이터 추출 가능

tt.query("sex =='male'")
```

```python
## 수치형 데이터는 ?
## 대소비교 < > <= >=
tt.query('fare>50')
```

```python
## 다중 조건인 경우는?
## 요금은 50 이상에 나이는 20 이상 다중조건
## and , or 조건 값이 달라짐
tt.query('fare>50 & age<20')
```

```python
# and 조건 , or 조건
tt.query('fare>50 | age<20')
```

```python
tt.query("embark_town in ['Cherbourg','Southampton']")
```

### 불린참값
- df[불린참값]는 True인 행만 골라주는 기능이야!

- 이 불린값은 주로 df['컬럼명'] > 값, ==, != 같은 조건식으로 만들어.
->  데이터 추출, 필터링, 조건부 분석에서 자주 씀

```python
tt[tt['sex'] == 'female']
```

```python
# 데이터프레임은 재구조화가 된다라는 거임

tt[(tt['sex']=='female')&(tt['fare']>50)][['survived','fare']]
```
